```python
import pandas as pd

df = pd.read_csv("../dataset/creditcard.csv")

print(df.shape)
df.head()
```

```python
df.info()
```

Class Distribution (Fraud vs Normal)

```python
import matplotlib.pyplot as plt

df['Class'].value_counts().plot(kind='bar')

plt.title("Fraud vs Normal Transactions")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
fraud_percentage = df['Class'].mean()*100
print("Fraud Percentage:", fraud_percentage)
```

```python
import seaborn as sns

sns.histplot(df['Amount'], bins=50)

plt.title("Transaction Amount Distribution")
plt.show()
```

```python
sns.boxplot(x='Class', y='Amount', data=df)

plt.title("Transaction Amount by Class")
plt.show()
```

```python
plt.hist(df['Time'], bins=50)

plt.title("Transaction Time Distribution")
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(12,10))

sns.heatmap(corr, cmap="coolwarm")

plt.title("Feature Correlation Matrix")
plt.show()
```

```python
fraud_corr = df.corr()['Class'].sort_values()

print(fraud_corr.head(10))
print(fraud_corr.tail(10))
```

```python
sns.scatterplot(x=df['V14'], y=df['V17'], hue=df['Class'])
plt.show()
```

```python
df['Class'].value_counts()
```

```python
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Normal")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X = df.drop("Class",axis=1)

X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:,0], X_pca[:,1], c=df["Class"], cmap="coolwarm", alpha=0.5)

plt.title("PCA Visualization of Fraud vs Normal")
plt.show()
```
